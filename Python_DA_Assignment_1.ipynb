```python
#2d array
temperatures = np.array([
    [22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9],
    [19.2, 22.5, 21.3, 24.0, 23.5, 22.8, 20.1]
])
print(temperatures)
```

```python
print("Shape:", temperatures.shape)
print("Data Type:", temperatures.dtype)
print("Total Elements:", temperatures.size)

# Each week
print("Week 1:", temperatures[0])
print("Week 2:", temperatures[1])

# Weekend (last 2 days of both weeks)
print("Weekend both weeks:\n", temperatures[:, -2:])
```

```python
marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5']
)
print(marks)
```

```python
# 1st rank using position
print("1st Rank:", marks.iloc[0])

# Top 3 using loc
print("Top 3:\n", marks.loc[['Rank1', 'Rank2', 'Rank3']])

# 3rd rank using iloc
print("3rd Rank:", marks.iloc[2])

# Boolean mask (>90)
print("Marks > 90:\n", marks[marks > 90])
```

```python
# Modifying Rank1
marks['Rank1'] = 100

# Remove last rank
marks = marks.drop('Rank5')

# Compute CGPA
cgpa = marks / 10

print("Updated Marks:\n", marks)
print("CGPA:\n", cgpa)
```

```python
transactions = pd.DataFrame({
    'TransactionID': [101,102,103,104,105,106,107,108,109,110],
    'ProductCategory': ['Electronics','Clothing','Electronics','Furniture','Clothing',
                        'Electronics','Furniture','Clothing','Furniture','Electronics'],
    'Region': ['North','South','North','East','West','North','East','West','South','North'],
    'Amount': [200,150,300,450,200,250,300,180,350,400]
})

print(transactions)
```

```python
# Head, Tail
print(transactions.head())
print(transactions.tail())
```

```python
# Shape
print("Shape:", transactions.shape)

# Columns
print("Columns:", transactions.columns)
```

```python
# Data types
print(transactions.dtypes)
```

```python
# Select columns
print(transactions[['ProductCategory', 'Amount']])
```

```python
# Last 3 columns
print(transactions.iloc[:, -3:])
```

```python
# Filter condition
filtered = transactions[(transactions['Region'] == 'North') & (transactions['Amount'] > 200)]
print(filtered)
```

```python
# Value counts
print(transactions['ProductCategory'].value_counts())
```

```python
# Unique regions
print(transactions['Region'].unique())
```

```python
# Group by Region
print(transactions.groupby('Region')['Amount'].mean())
```

```python
transactions.loc[transactions['TransactionID'] == 102, 'Amount'] = 165
```

```python
#add discount
transactions['Discount'] = transactions['Amount'] * 0.10
```

```python
#remove 109
transactions = transactions[transactions['TransactionID'] != 109]
```

```python
#del discounrt
transactions = transactions.drop(columns=['Discount'])
```
